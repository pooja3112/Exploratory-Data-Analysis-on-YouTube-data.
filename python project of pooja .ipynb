```python
#To create a new column named 'published_month' using the 'published_at' column and extract only the months,used the dt.month attribute of the pandas datetime objects.

#'df' with the 'published_at' column of datetime type with the method dt.month

df['published_month']=df['published_at'].dt.month

# Print the updated DataFrame with the new 'published_month' column
print(df['published_month'])
```

```python
#To create a new column named 'published_year' using the 'published_at' column and extract only the years,used the dt.year attribute of the pandas datetime objects.

#'df' with the 'published_at' column of datetime type with the method dt.year

df['published_year']=df['published_at'].dt.year

# Print the updated DataFrame with the new 'published_year' column

print(df['published_year'])
```

```python
#To create a new column named 'published_date' using the 'published_at' column and extract only the date,used the dt.date attribute of the pandas datetime objects.

#'df' with the 'published_at' column of datetime type with the method dt.date
df['published_date']=df['published_at'].dt.date

# Print the updated DataFrame with the new 'published_date' column
print(df['published_date'])
```
